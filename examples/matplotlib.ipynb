```python
%%capture
import lantern as l
```

```python
df = l.bar.sample()
```

```python
col1 = df.columns[0]
col2 = df.columns[1]
```

```python
l.plot(df[[col1, col2]], size=(12,5), subplot=True)
```

```python
f = l.figure()
f.line(df[df.columns[:1]])
f.bar(df[df.columns[1:2]])
f.area(df[df.columns[1:2]])
f.show()
```

```python
f = l.figure()
f.line(df[df.columns[:1]], y_axis='right')
f.bar(df[df.columns[1:2]])
f.area(df[df.columns[1:2]], y_axis='right')
f.show(xticks=False)
```
